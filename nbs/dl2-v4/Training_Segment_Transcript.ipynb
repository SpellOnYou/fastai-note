```
def segmen_subsents(n, _out=unlist):
    if n ==0:
        return _out
    else:
        nested_out = [segmenter.segment(_inp) for _inp in _out]
        _unlist = [j for i in nested_out for j in i]
        print(n, len(_unlist))
        return segmen_subsents(n-1 , _out=_unlist)
```
